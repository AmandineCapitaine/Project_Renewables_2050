```python
# Moyenne de chaque heure de la journée sur un an
# Vecteur de 24 (nb d'heures)

def Moyenne_jour(profil):
    heures, jours = np.shape(profil)
    moy_j = np.zeros(heures)
    for j in range(jours):
        for h in range(heures):
            moy_j[h] += profil[h,j]
    moy_j = moy_j/jours
    return moy_j


# Affichage du profil de production d'une journée calculé avec les données d'une année entière

def Profils_journaliers(profil, scenario, annee, verbose=False):
    
    profil_jour = np.zeros((24,365))
    for j in range(362):
        for i in range(24):
            profil_jour[i,j] += [profil[(i+1)+j*24]]
            
    moy_jour = Moyenne_jour(profil_jour)
    plt.plot(moy_jour, label=annee)
    plt.title("Profil journalier de la demande : " + scenario)
    plt.xlabel("Heures")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
    
    if verbose :
        return profil_jour
```

```python
pademe = Profils_journaliers(ADEME['demande'],'ADEME', '2050', True)
```

```python
def Variance_profil(profil):
    var = np.zeros(24)
    for i in range(24):
        var[i] = np.var(profil[i][:])
    return var

var_ademe = Variance_profil(pademe)
plt.plot(var_ademe)
plt.title('Variance journalière demande ADEME')
plt.show()
```

```python
prte = Profils_journaliers(RTE['demande'],'RTE', '2050', True)
```

```python
var_rte = Variance_profil(prte)
plt.plot(var_rte)
plt.title('Variance journalière demande RTE')
plt.show()
```

```python
pnega = Profils_journaliers(Negawatt['demande'],'Negawatt', '2050', True)
```

```python
var_nega = Variance_profil(pnega)
plt.plot(var_nega)
plt.title('Variance journalière demande Negawatt')
plt.show()
```
